```python
import polars as pl
df_subm_9999_725 = pl.read_csv('./resnet_warmup/submission.csv')
df_subm_760_721 = pl.read_csv('./seq2seq/submission.csv')
df_subm_42_722 = pl.read_csv('./seq2seq1/submission.csv')
df_subm4_521_727 = pl.read_csv('./seq2seq2_attention/submission.csv')
```

```python
columns_to_add =df_subm_9999_725.columns[1:]  # 假设第一列是'id'，我们不包括这列
result = df_subm_9999_725.select(columns_to_add) * 0.35 + df_subm_760_721.select(columns_to_add) * 0.125 + df_subm_42_722.select(columns_to_add) * 0.125 + df_subm4_521_727.select(columns_to_add) * 0.4
```

```python
result
```

```python
result = pl.concat([df_subm_9999_725.select(df_subm_9999_725.columns[0]), result], how='horizontal')
```

```python
result.write_csv("submission.csv")
```
